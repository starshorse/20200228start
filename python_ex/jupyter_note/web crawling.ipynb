### List 요소를 4개로 나누어 다시 배열로 만든다.

```python
# List 요소를 4개로 나누어 다시 배열로 만든다.
def list_chuck(arr, n):
    return [arr[i: i + n] for i in range(0, len(arr), n)]

items = driver.find_element('xpath','//*[@id="boxInfluentialInvestors"]/div[2]/div[1]/table')
#print( items.text )
list_items_info = items.text.splitlines() # 텍스트를 라인별로 잘라서 Array 로 만들어 준다.
#print( list_items_info )

del list_items_info[:2]    # 앞에 필요업는 2개 배열요서 제거 
#print( list_items_info )

forEn = list_chuck( list_items_info, 4 )
print( forEn )
```

```python
import pandas as pd

df = pd.DataFrame(forEn)
df
```

```python
df.columns =['종목명','금액','수량','등락']
df.info()
```

### List 에서 홀수째 열만 뽑아 낸다.

```python
mList = [ x for x in range(1, df['종목명'].count(), 2 )]
mList 
```

```python
df['금액'] = df['금액'].str.replace(',','')
df['수량'] = df['수량'].str.replace(',','')
# df.drop( index..)
df = df.drop(mList)
df
```

```python
df['금액'] = df['금액'].astype('int')
df['수량'] = df['수량'].astype('int')
df.dtypes
```

### 필요업는 항목의 제목으로 제거하기 

```python
df = df[~df['종목명'].str.contains('TIGER')]
df = df[~df['종목명'].str.contains('KODEX')]
df.sort_values( by ='금액')
```
