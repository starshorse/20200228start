```python
from selenium import webdriver
from selenium.webdriver.common.by import By
driver = webdriver.Chrome('./chromedriver')
```

```python
driver.get('https://finance.daum.net/domestic/influential_investors')
```

```python
items = driver.find_element('class name','box_contents')
print( items.text )
```

```python
def list_chuck(arr, n):
    return [arr[i: i + n] for i in range(0, len(arr), n)]

items = driver.find_element('xpath','//*[@id="boxInfluentialInvestors"]/div[2]/div[1]/table')
# print( items.text )

list_items_info = items.text.splitlines()
# print( list_items_info )

del list_items_info[:2]
# print( list_items_info )

forEn = list_chuck( list_items_info, 4 )

print( forEn )
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(forEn)
```

```python
df.columns =['종목명','금액','수량','등락']
df.info()
```

```python
mList = [ x for x in range(1, df['종목명'].count(), 2 )]
mList 
```

```python
df['금액'] = df['금액'].str.replace(',','')
df['수량'] = df['수량'].str.replace(',','')
df = df.drop(mList)
df
```

```python
df['금액'] = df['금액'].astype('int')
```

```python
df['수량'] = df['수량'].astype('int')
```

```python
df.dtypes
```

```python
df = df[~df['종목명'].str.contains('TIGER')]
df = df[~df['종목명'].str.contains('KODEX')]
df.sort_values( by ='금액')
```

```python
df.index = list( range(0, df['종목명'].count())) 
```
